```python
import pandas as pd
```

```python
bible = pd.read_table('./corpora/kjvdat.txt', sep='|', header=None, names=['book', 'verse', 'phrase', 'text'])
```

```python
bible['text'] = bible['text'] + ' eos'
```

```python
txts = bible.groupby('book')['text'].apply(lambda s: f"".join(s)).reset_index()
```

```python
SELECTED_BOOKS = {'Gen': 'genesis',
                  'Exo': 'exodus',
                  'Lev':  'leviticus',
                  'Deu': 'deuteronomy',
                  'Sa1': '1_samuel',
                  'Sa2':  '2_samuel',
                  'Kg1':  '1_kings',
                  'Kg2':  '2_kings',
                  'Est':  'esther',
                  'Dan':  'daniel',
                  'Ch1':  '1_chronicles',
                  'Ch2':  '2_chronicles',
                  'Ecc':  'ecclesiastes'}
```

```python
txts['book'] = txts['book'].map(SELECTED_BOOKS)
```

```python
txts = txts.dropna()
```

```python
txts = txts.set_index('book').squeeze().to_dict()
```

```python
for k in txts:
    txts[k] = txts[k].replace('~', '')
    txts[k] = " ".join(txts[k].split())
    txts[k] = txts[k].lower()

    with open(f'./corpora/english/{k}.txt', 'w') as f:
        f.write(txts[k])
```
